### Managed vs external tables
[video](https://youtu.be/761SQ9Hxbic?list=PLeo1K3hjS3uu7dS3Cx0X796sWUjjBFCuV&t=5988)

**Any tables will have two components: content and metadata**.

if it is a **managed** table databrick will manage both.

metadata is inside of the databrick, but the actual data can be outside of the databrick, it can be in AWS S3, etc...


### Managed Tables
- Databricks manages both metadata and actual sotrage of data
- Multi-tool access: Difficult
- Data governance: Unity catalog controls everything
- Use case: Quick analytics, fully governed data

### External Tables
- Databricks manages only metadata
- Multi-tool access: Easy
- Data governance: Flexible but requires discipline
- Use case: Data shared with other tools or pre-existing datasets
