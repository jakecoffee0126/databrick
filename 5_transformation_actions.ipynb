### Transformations:
A transformation is an operation that defines a new daaset from an existing one but DOES NOT execute immediately.
- select, withColumn, filter, where, groupBy, join, sort, orderBy, etc.....

### Actions:
An action is an operation that triggers the actual execution of all pending transformations and returns a result to the driver program or writes data to storage.
- show(), count(), toPandas(), collect(), write(), save(), etc.....

### Benefit with Lazy Evaluation
- Query Optimization 
- Avoid unecessary work
- Fault Tolerant

```python
from pyspark.sql import functions as F

df = spark.table("workspace.default.movies") #transformation
df_narrow = df.select("title", "studio", "imdb_rating").filter(F.col("release_year") >= 2010) #transformation

#in spark, line#4 will NOT be executed when you run line #4, it will be executed when call the action, like display(df_narrow) or df_narrow.show()

# Action - now everyting executes
display(df_narrow) # action - triggers all aobe transformation
df_narrow.show()# action - triggers all aobe transformation
```
