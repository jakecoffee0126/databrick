```python
df = spark.table("workspace.default.movies")
df.show(3)
df.printSchema()
```

```python
from pyspark.sql import functions as F
df_narrow = df.select("title", "studio", "imdb_rating").filter(F.col("release_year") > 2010)
display(df_narrow)
```
